```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
```

```python
q = QuantumRegister(2)
qq = QuantumRegister(6)
c = ClassicalRegister(2)
Bell = QuantumCircuit(q)
qc = QuantumCircuit(qq, c)
```

```python
# Bell State 

Bell.reset(0)
Bell.reset(1)

Bell.h(0)
Bell.cx(0, 1)

Bell.draw()
```

```python
qc.reset(0)
qc.reset(1)
qc.reset(2)
qc.reset(3)
qc.reset(4)
qc.reset(5)

qc.h(0) # Hadamard of the Bell State

# P R E P A R A T I O N 
qc.cx(0, 1)
qc.cx(0, 2)
qc.cx(3, 4)
qc.cx(3, 5)
qc.h(0)
qc.h(1)
qc.h(2)

# E R R O R S
qc.z(0) # error: Z gate. It flips the | + >  and | - > states. Sign- flip error
qc.x(3) # error: X gate. It flips | 0 > state to | 1 > state, and vice versa. Bit flip error

# C O R R E C T I O N
qc.cx(3, 4) # BIT FLIP
qc.cx(3, 5)
qc.ccx(5, 4, 3)

qc.h(0) # SIGN FLIP
qc.h(1)
qc.h(2)
qc.cx(0, 1)
qc.cx(0, 2)
qc.ccx(2, 1, 0)

qc.cx(0, 3) # CNOT of the Bell State

qc.draw()
```

```python
# M E A S U R E M E N T
qc.measure(0, 0)
qc.measure(3, 0)

qc.draw()
```

1. Bell State: I have represented the Bell State at the beggining, in a separate circuit. Bell State is the maximal example of quantum entanglement. 
2. Right before the CNOT, in my main circuit, I add the arbitrary error gate. I decided to add some qubits in order to operate two different error corrections, the Z and the X gates are my errors.
3. I operate the BIT FLIP and the SIGN FLIP correction, respectively for X and Z gate. I used these algorihms because Z(| + >) = | - > and I had to get | + > back, same for X(| 0 >)= | 1 > where I had to get back the | 9 > state.
4. I conclude by measuring the circuit, both in the Circuit composer of IBMQ Experience and in the code.
